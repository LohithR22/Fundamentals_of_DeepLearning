```python
import math
```

```python
def basic_sigmoid(x):
    s = 1/(1+math.exp(-x))
    return s
```

```python
x = basic_sigmoid(-1)
x
```

```python
import numpy as np

x = np.array([1,2,3])
print(np.exp(x))
```

```python
def np_sigmoid(x):
    s = 1/(1+np.exp(-x))
    return s
```

```python
x = np_sigmoid(np.array([-1,2,3]))
x
```

```python
def grad_sig(x):
    sig = np_sigmoid(x)
    grad = sig*(1-sig)
    return grad

x = grad_sig(np.array([-1,2,3]))
x
```

```python
image = np.random.rand(3, 3, 2)
image
```

```python
def dim_1D(img):
    img = img.reshape(img.shape[0]*img.shape[1]*img.shape[2], 1)
    return img

image = dim_1D(image)
image
```

```python
def norm_row(img):
    x_norm = np.linalg.norm(img, axis=1, keepdims = True)
    img  = img/x_norm
    return img

image = norm_row(image)
image
```

```python
img1 = np.array([[0,1000,4], 
                 [2,6,4]])
img1 = norm_row(img1)
img1
```
